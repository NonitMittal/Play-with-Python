# Creating Team List

```python
import numpy as np
import pandas as pd

data = pd.read_csv('C:\\Users\\HP\\Desktop\\ipl2017 maths assignment.csv', )
teamsl = list(data)
teams = []
for x in range(1,9):
    teams.append(teamsl[x])
print("\n", teams)
```

# Creating Matrix

```python
mat = pd.read_csv('C:\\Users\\HP\\Desktop\\ipl2017 maths assignment.csv')
print(mat, "\n\n\n")

n = 8

mat = np.matrix(mat)
mat2 = []
for x in range(8):
    mat2.append([])
    for y in range(1, 9):
        mat2[x].append(mat[x , y])
        
mat = np.matrix(mat2)
print(mat)
```

# Finding Dominant Eigen Value and Eigen Vector via Power Method  

```python
import numpy as np
unit = np.ones((n,1))
m = np.matrix(unit)

b = True
itr = 0
while b:
    itr += 1
    y = 0
    p = mat@m
    x = p.max()
    p = p / x

    for i in range(n):
        if round(p.item(i), 4) == round(m.item(i), 4):
            y += 1

    print("\n\n Iteration Number: ", itr, "\n")

    if(y == n):
         b=False
        
    print("Eigen Value: ",x)
    print("Eigen Vector: \n",p)        
    
    m = p
```

# Ranking

```python
rank = []
rankus = []
for i in range(n):
    rank.append(p.item(i))
    rankus.append(p.item(i))
rank.sort()

for i in range(n):
    print(teams[rankus.index(rank[-(i+1)])], "\n")

del rank
del rankus
```
